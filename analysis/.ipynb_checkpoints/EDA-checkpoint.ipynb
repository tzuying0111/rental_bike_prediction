# Univariate Analysis

```python
df.describe()
```

```python
def plt_hist(col):
    f, axes = plt.subplots(1, 2, figsize=(10, 4))
    f.tight_layout()
    sns.distplot(df[col], ax=axes[0])
    sns.boxplot(df[col], orient='v', ax=axes[1])
    plt.show()
```

```python
for col in non_str_col:
    plt_hist(col)
```

```python
cat_col
```

```python
cat_df = df.select_dtypes(include=['object']).copy()
```

```python
def plt_bar(col):
    plt_df = cat_df[col].value_counts()
    plt.title(col)
    sns.barplot(plt_df.index, plt_df.values)
    plt.show()
```

```python
for col in cat_col:
    plt_bar(col)
```
